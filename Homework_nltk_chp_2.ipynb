```python
#1.
def generate_model(text, n):
    text_fd=nltk.FreqDist(text)
    text_fd_common=text_fd.most_common(n)
    rand_words = [word for word,index in text_fd_common]
    return rand_words
```

```python
# Write a function gematria() that sums the numerical values of the letters of a word, according to the letter values in letter_vals:
def gematria(text):
  x = re.findall("[A-z]",text.lower())
  y = [alpha_value(i) for i in x]
  #print(sum(y),y)
  return sum(y)
gematria('Dont,]')

def alpha_value(ch):
  letter_vals = {'a':1, 'b':2, 'c':3, 'd':4, 'e':5, 'f':80, 'g':3, 'h':8, 'i':10, 'j':10, 'k':20, 'l':30, 'm':40, 'n':50, 'o':70, 'p':80, 'q':100, 'r':200, 's':300, 't':400, 'u':6, 'v':6, 'w':800, 'x':60, 'y':10, 'z':7}
  if ch in letter_vals.keys():
    return letter_vals[ch]
  else:
    return 0

```

```python
#Read the Wikipedia entry on Soundex. Implement this algorithm in Python. 50 pts

def soundex(given_name):
  code = ''
  grp = {1 :['a', 'e', 'i', 'o', 'u', 'y', 'h', 'w'],
  2 : ['b', 'f', 'p', 'v'],
  3 : ['c', 'g', 'j', 'k', 'q', 's', 'x', 'z'],
  4 : ['d', 't'],
  5 : ['l'],
  6 : ['m','n'],
  7 : ['r']}
  given_name.lower()
  if given_name[0] == given_name[1]:
    code = given_name[0].upper()
  else:
    code = given_name[0].upper()
    list(given_name).remove(code.lower())
    for i in given_name:
      if i in grp[1]:
        list(given_name).remove(i)
      if i in grp[2]:
        code = code+'1'
      if i in grp[3]:
        code = code +'2'
      if i in grp[4]:
        code = code+'3'
      if i in grp[5]:
        code = code+'4'
      if i in grp[6]:
        code = code+'5'
      if i in grp[7]:
        code = code+'6'
    set(code)
    if len(code) < 4:
      num = 4 - len(code)
      code = code+zero(num)
  return code


def zero(num):
  return '0'*num

soundex('Rupert')

#Using this algorithm, both "Robert" and "Rupert" return the same string "R163" while "Rubin" yields "R150".
# "Ashcraft" and "Ashcroft" both yield "A261". "Tymczak" yields "T522" not "T520" (the chars 'z' and 'k' in the name are coded as 2 twice since a vowel lies in between them).
# "Pfister" yields "P236" not "P123" (the first two letters have the same number and are coded once as 'P'), and "Honeyman" yields "H555".
```
