```python
#1.

def rotate_word(str1,num):
  new_str = ""
  for i in str1:
    new_str = new_str + chr(ord(i)+num)
  return new_str

rotate_word("Ab",1)
```

```python
#c 

def bigrams(long_string):
  all_words = long_string.split()
  bigrams_list = [(all_words[i],all_words[i+1]) for i in range(0,len(all_words)-1)]
  return bigrams_list

bigrams(long_string)

```
