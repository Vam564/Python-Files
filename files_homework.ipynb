```python
#Part 1
#1.
import os
from pathlib import Path
os.mkdir('/class_files')
os.mkdir('/class_files/texts')
```

```python
#2
os.chdir('/class_files')
os.path.abspath('texts')
```

```python
#3
import shutil
import glob
source_dir = '/home'
target_dir = '/class_files/texts'
    
file_names = os.listdir(source_dir)
    
for file_name in file_names:
    shutil.move(os.path.join(source_dir, file_name), target_dir)
```

```python
#4
os.chdir('/class_files/texts')
g = os.listdir(target_dir)
for file_s in g:
  print(os.path.abspath(file_s))
for x in glob.glob('*.txt'):
  print(x)
```

```python
#Part2
#1
z = glob.glob('*.txt')
file_1 = open(z[0],'r')
file_2 = open(z[1],'r')
```

```python
#2
file_1_data = file_1.read()
file_2_data = file_2.read()
print(file_1_data)
print(file_2_data)
file_1.close()
file_2.close()
```

```python
#3
print(len(file_1_data),'File 1 length')
print(len(file_2_data),'File 2 length')
```

```python
#4
file_1_list = file_1_data.split()
file_2_list = file_2_data.split()
print(file_1_list)
print(file_2_list)
```

```python
#5
print(len(file_1_list),"File 1 list length")
print(len(file_2_list),"File 2 list length")
```

```python
#bonus question

for i,words in enumerate(file_1_list):
  file_1_list[i] = words.upper()
for i,words in enumerate(file_2_list):
  file_2_list[i] = words.upper()
  #words changed to Uppercase
updated_file_1_data = ' '.join([str(word) for word in file_1_list])
updated_file_2_data = ' '.join([str(word) for word in file_2_list])
file_1 = open(z[0],'w')
file_1.write(updated_file_1_data)
file_2 = open(z[1],'w')
file_2.write(updated_file_2_data)
```
